```python
tb = pd.read_csv('tb.csv')
```

```python
tb = tb.melt(['country','year'])
```

```python
tb.head()
```

```python
tb['sex'] = tb['variable'].apply(lambda s: s[0])
tb['age'] = tb['variable'].apply(lambda s: s[1:])
```

```python
#tb.head()
tb.age.unique()
```

```python
tb.age = tb.age.map({'04':'0-4', '514':'5-14', '014':'0-14', '1524':'15-24', '2534':'25-34', '3544':'35-44', '4554':'45-54', '5564':'55-64', '65':'65+',
       'u':np.nan})
```

```python
tb.head()
```

```python
tb.drop('variable',axis=1,inplace=True)
```

```python
tb.head()
```

```python
tb_new = tb.groupby(['country','year','sex','age'])['value'].sum().reset_index().sort_values(by = ['country','year','age'])
tb_new.head()
```

```python
vx.head(10)
```

```python
tb_new = tb_new[['country','year','age','sex','value']]
```

```python
tb_new = tb_new.rename(columns = {'value':'cases'})
```

```python
tb_new.head()
```

```python
tb_new = tb_new.dropna(subset=['cases'])
```

```python
tb_new = tb_new.reset_index()
tb_new.head()
```
