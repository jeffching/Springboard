#### Can you count something interesting?

```python
data.sex.value_counts()
```

```python
data.sex.value_counts()[1]
```

```python
data.sex.value_counts()[2]
```

```python
print("Male % applied for loan:", round(data.sex.value_counts()[1]/len(data),4))
print("Female % applied for loan:", round(data.sex.value_counts()[2]/len(data),4))
```

```python
data.sex.plot(kind='hist',bins=10,density=True)
```

Interestingly, there is an imbalance of more female loan applicants than male (+20%)

```python
data.sex.value_counts()[2]
```

```python
#data.plot.scatter("year","population")
data.query('sex == 1').plot.scatter("defaulted","limit_bal")
```

```python
data.query('sex == 2').plot.scatter("defaulted","limit_bal")
```

We see a higher number of females who default at a higher loan rate than males (peak of 750k vs peak of 600k)

```python
data.columns
```

```python
data_defaulted = data.query('defaulted == 1')
data_nodefault = data.query('defaulted == 0')
data_defaulted.query('pay_amt1 < bill_amt1').plot.scatter('pay_amt1','bill_amt1')
data_nodefault.query('pay_amt1 < bill_amt1').plot.scatter('pay_amt1','bill_amt1')
#data_defaulted.plot.scatter('pay_amt1','bill_amt1')
```

```python
data_defaulted.query('pay_amt1 >= bill_amt1').plot.scatter('pay_amt1','bill_amt1')
data_nodefault.query('pay_amt1 >= bill_amt1').plot.scatter('pay_amt1','bill_amt1')
```

```python
data_defaulted.query('pay_amt3 >= bill_amt3').plot.scatter('pay_amt3','bill_amt3')
data_nodefault.query('pay_amt3 >= bill_amt3').plot.scatter('pay_amt3','bill_amt3')
```

```python
data_defaulted.query('pay_amt6 >= bill_amt6').plot.scatter('pay_amt6','bill_amt6')
data_nodefault.query('pay_amt6 >= bill_amt6').plot.scatter('pay_amt6','bill_amt6')
```

```python
We see borrowers who did not default on their loan were generally consistent with their debt repayment (pay_amt* >= bill_amt*)
```
