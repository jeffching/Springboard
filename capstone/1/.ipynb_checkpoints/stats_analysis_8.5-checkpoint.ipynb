

**pay_**: repayment status.  pay_1 = the repayment status in September, 2005; pay_2 = the repayment status in August, 2005; . . .;pay_6 = the repayment status in April, 2005. In both cases of default and non-defaulted borrowers, the payment and bill amount status appear highly correlated.  Instead, the actual payment amount tracked for each bill repayment period *(pay_amt*)* may have higher feature importance than the repayment status or bill amount. This suggests that we can strongly consider excluding the highly correlated features from our statistical analysis as they would not add value to the modeling stage. We see that repayment status, while there shows a strong correlation, it is not high enough (~50-80%) that we may consider excluding at this time.

We can use Bayesian Inference to model the distribution of individual repayment statuses and also be able to capture our uncertainty about that distribution so we can better capture the range of values we might see. 

```python
default = df2.pay_1[df2.defaulted == 1]
no_default = df2.pay_1[df2.defaulted  == 0]
n_default = len(default)
n_no_default = len(no_default)
```

```python
default.value_counts()
```

```python
#Initial parameter estimation
alpha_est = np.mean(default)**2 / np.var(default)
beta_est = np.var(default) / np.mean(default)
alpha_est, beta_est
```

```python
#Empirical distribution
_ = plt.hist(no_default, bins=10, alpha=0.5, label='not defaulted clients')
_ = plt.hist(default, bins=10, alpha=0.5, label='defaulted clients')
_ = plt.xlabel('Repayment Status')
_ = plt.ylabel('Frequency')
_ = plt.legend()
```

```python
#Initial simulation
seed(123)
default_model_rvs = gamma(alpha_est, scale=beta_est).rvs(n_default)
```

```python
_ = plt.hist(default_model_rvs, alpha=0.5, label='simulated')
_ = plt.hist(default, alpha=0.5, label='observed')
_ = plt.xlabel('Charge amount')
_ = plt.ylabel('Frequency')
_ = plt.legend()
```

```python
#Create a PYMC3 model
# PyMC3 Gamma seems to use rate = 1/beta
rate_est = 1/beta_est
# Initial parameter estimates we'll use below
alpha_est, rate_est
```

```python
with pm.Model() as model:
    #Create your Gamma distribution with your alpha_ and rate_ stochastic variables and the observed data.
    gamma = pm.Gamma('gamma',alpha=alpha_est,beta=beta_est)
    #Perform 10000 draws.
    trace = pm.sample(10000,tune=) #use NUTS for continuous model sampling
```

```python
t = pm.stats.hpd(trace)
```

```python
pm.plots.plot_posterior(trace,kind="hist")
```

```python
pm.traceplot(trace)
```

```python
from scipy.stats import gamma
simulated_data = gamma.rvs(alpha_est,scale = beta_est,size = len(default))
```

```python
seed(123)
```

```python
_ = plt.hist(simulated_data, bins=10, label='simulated data')
_ = plt.hist(default, bins=10, alpha=0.8, label='observed data')
_ = plt.xlabel('pay_1')
_ = plt.ylabel('Frequency')
_ = plt.legend()
```

In both the Gamma distribution and simulated distribution, we see that the defaulters were either not behind or only one payment month behind. In comparison to non-default individuals, none of the other variables appear to observe a significant difference in behavior from value counts. Additionally because we have an underrepresented dataset, this would need to be taken into account prior to the modeling stage.
